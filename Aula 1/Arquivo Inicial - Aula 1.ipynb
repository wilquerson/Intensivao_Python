# Automação de Sistemas e Processos com Python

### Desafio:

Todos os dias, o nosso sistema atualiza as vendas do dia anterior.
O seu trabalho diário, como analista, é enviar um e-mail para a diretoria, assim que começar a trabalhar, com o faturamento e a quantidade de produtos vendidos no dia anterior

E-mail da diretoria: seugmail+diretoria@gmail.com<br>
Local onde o sistema disponibiliza as vendas do dia anterior: https://drive.google.com/drive/folders/149xknr9JvrlEnhNWO49zPcw0PW5icxga?usp=sharing

Para resolver isso, vamos usar o pyautogui, uma biblioteca de automação de comandos do mouse e do teclado

```python
#!pip install pyautogui
import pyautogui as pya
import pyperclip as pyp
import time

#pyautogui.click -> clicar
#pyautogui.write -> escrever
#pyautogui.press -> pressionar
#pyautogui.hotkey -> atalho

#pyautogui.PAUSE = 1 #pausa no codigo

# Passo 1: Entrar no sistema da empresa (no link do drive)
pya.hotkey('ctrl', 't')
pyp.copy('https://drive.google.com/drive/folders/149xknr9JvrlEnhNWO49zPcw0PW5icxga?usp=sharing')
pya.hotkey('ctrl', 'v')
pya.press('enter')

time.sleep(3) #pausa customizada na linha

# Passo 2: Navegar até o local do relatório (entrar na pasta Exportar)
pya.click(x=389, y=314, clicks=2)
time.sleep(2)

# Passo 3: Exportar o relatório (fazer o download)
pya.click(x=389, y=314)
pya.click(x=1160, y=184)
pya.click(x=976, y=589)
time.sleep(8)
pya.press('enter')
```

### Vamos agora ler o arquivo baixado para pegar os indicadores

- Faturamento
- Quantidade de Produtos

```python
# Passo 4: Calcular os indicadores (faturamento e quantidade de produtos)
import pandas as pd

tabela = pd.read_excel(r'C:\Users\wilqu\Documents\Intensivão Python\Vendas - Dez.xlsx') #r - ler exatamente o que está escritono link
display(tabela)

faturamento = tabela["Valor Final"].sum()
quantidade = tabela["Quantidade"].sum()
print(faturamento)
print(quantidade)
```

### Vamos agora enviar um e-mail pelo gmail

```python
# Passo 5: Enviar um email para a diretoria

#abrir aba e entrar no email
pya.hotkey('ctrl', 't')
pyp.copy('https://mail.google.com/mail/u/0/?tab=rm&ogbl#inbox')
pya.hotkey('ctrl', 'v')
pya.press('enter')
time.sleep(3)

#clicar em escrever
pya.click(x=109, y=196)
time.sleep(2)

#inserir o email
pya.write('wilquerson@hotmail.com')
pya.hotkey('tab')
pya.hotkey('tab')

#assunto
pyp.copy('Relatório de vendas')
pya.hotkey('ctrl', 'v')
pya.hotkey('tab')

#corpo do email
texto = f'''
Prezados senhores, bom dia.

O faturamento é R$ {faturamento:,.2f}
A quantidade é {quantidade:,}

At.te,
Wilquerson'''

pyp.copy(texto)
pya.hotkey('ctrl', 'v')
pya.hotkey('ctrl', 'enter')
```
